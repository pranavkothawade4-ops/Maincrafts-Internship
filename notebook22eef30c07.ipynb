```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/kaggle/input/datasets/pranavkothawade2005/task1main/student-mat.csv", sep=";")

# Display first 5 rows
df.head()

```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
df.shape
```

```python
df["G3"].mean()
```

```python
(df["G3"] > 15).sum()
```

```python
df["studytime"].corr(df["G3"])
```

```python
df.groupby("sex")["G3"].mean()
```

```python
plt.figure()
plt.hist(df["G3"], bins=10)
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.title("Distribution of Final Grades")
plt.show()
```

```python
plt.figure()
plt.scatter(df["studytime"], df["G3"])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade")
plt.show()
```

```python
avg_scores = df.groupby("sex")["G3"].mean()

plt.figure()
plt.bar(avg_scores.index, avg_scores.values)
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.title("Average Grade by Gender")
plt.show()
```
